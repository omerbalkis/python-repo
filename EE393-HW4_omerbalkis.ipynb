This program plots a third order polynomial **y=ax^3 + bx^2 + cx + d** and it's first derivative. In this program, we are assuming that X in the interval -1 and 1 for y and y'. 
Written by Hilmi Ömer Balkış - S009741

```python
def ourFunction (x,a,b,c,d):
    generalSolution = a*x**3 + b*x**2 + c*x + d
    return generalSolution

def ourFuntionsDerivative (x,a,b,c):
    generalSolution = 3*a*x**2 + b*x*2 + c
    return generalSolution
```

Two methods which above takes inputs from user and find results according to X values in range -1 and 1. I create a library for more cleaner code in main class.

```python
import  matplotlib.pyplot as plt
import  myGraphLibrary as lb

#I fix my x values between -1 and 1
x = [-1, -0.5, 0, 0.5, 1]

#In this part, i asked float inputs from user to draw graph.
a = float(input("a :"))
b = float(input("b :"))
c = float(input("c :"))
d = float(input("d :"))

#I created two empty lists as y1 and y2 and i calculate their values for every x then append to lists.
y1 = []
y2 = []

for i in x:
    y1.append(lb.ourFunction(i, a, b, c, d))
    y2.append(lb.ourFuntionsDerivative(i, a, b, c))

#In this part, by using matplotlib.pyplot library i drew our graph.
plt.plot(x, y1, color='blue', marker='o', linestyle='dashed', linewidth=2, markersize=10)
plt.plot(x, y2, color='red', marker='o', linestyle='dashed', linewidth=2, markersize=10)
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('EE 393 - HW 4')
plt.grid(True)
plt.show()
```
